```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd

#Loading the dataset
df=pd.read_csv("Students.csv")

#finding Total Marks
df['total']=df[['Maths','Science','English']].sum(axis=1)

print(df)
print()

#xlabel,ylabel,title
plt.xlabel("Student Names")
plt.ylabel("Total Marks")
plt.title("STUDENT TOTAL MARKS IN ALL SUBJECTS",color='red')

#Creating bar chart
plt.bar(df['Name'],df['total'],color='green',edgecolor='red')

#display the output
plt.show()


```

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#Loading Dataset
df=pd.read_csv("Students.csv")

#Finding Average for all subjects
mat_avg=df['Maths'].mean()
sci_avg=df['Science'].mean()
eng_avg=df['English'].mean()

sub_avg=[mat_avg,sci_avg,eng_avg]

#Creating pie chart
plt.pie(sub_avg,autopct='%1.2f%%',explode=[0.02,0.02,0.02])

#title
plt.title(" SUBJECT-WISE AVERAGE MARKS",color='green')

#axis
plt.axis('equal')

#legend
plt.legend(df.columns[1:],bbox_to_anchor=(1,1))

#Display the Pie chart
plt.show()

```

```python
#3. Movie Ratings Visualization

#Goal: Compare ratings of 5 movies from different platforms (IMDB, Rotten Tomatoes, etc.).

#Tasks:

#Use grouped bar charts.

#Add legend and x-ticks with movie names.

#Skills: Grouped bar chart,Â tickÂ rotation.
```

```python
#importing libaries
import matplotlib.pyplot as plt
import numpy as np

#data
movies = ['RRR', 'Pushpa', 'Salaar', 'Sarkaru Vaari Paata', 'Dasara']
imdb = [8.0, 7.6, 7.2, 6.8, 7.4]
rotten = [92/10, 85/10, 80/10, 75/10, 82/10]
bookmyshow = [9.2, 8.9, 8.5, 8.0, 8.6]
google = [94/10, 90/10, 88/10, 85/10, 89/10]
audience = [9.0, 8.8, 8.2, 7.9, 8.3]
x=np.arange(len(movies))
width=0.15

#creating group bar chart
plt.bar(x - 2*width, imdb, width, label='IMDB', color='deeppink')
plt.bar(x -1* width, rotten, width, label='Rotten Tomatoes', color='purple')
plt.bar(x, bookmyshow, width, label='BookMyShow', color='orange')
plt.bar(x +1* width, google, width, label='Google Users', color='teal')
plt.bar(x + 2*width, audience, width, label='Audience', color='crimson')

#xlabel,ylabel,title,xticks
plt.xlabel("Movies")
plt.ylabel("Ratings")
plt.title(" MOVIES RATINGS OVER DIFFERENT PLATFORAMS",color='#3C096C')
plt.xticks(x,movies,rotation=30)

#legend
plt.legend(bbox_to_anchor=(1,1))

#displaying
plt.show()
```

```python
#4. Boxplot of Exam Scores

#Goal: Show the distribution of scores across multiple subjects.

#Skills: Boxplots, identifyingÂ outliers.
```

```python
#importing llibraries
import matplotlib.pyplot as plt
import pandas as pd

#data
d={
    'maths' :[85, 78, 92, 88, 76, 95, 89, 84, 90, 73, 99, 91, 87, 80, 77],
    'science' :[80, 75, 89, 84, 70, 94, 88, 82, 86, 69, 95, 90, 85, 78, 72],
    'english' :[78, 82, 88, 90, 84, 76, 85, 92, 81, 80, 83, 89, 91, 87, 79],
    'history' : [70, 72, 68, 75, 80, 77, 82, 85, 73, 69, 74, 81, 78, 83, 76],
    'computers' : [90, 94, 88, 92, 86, 96, 98, 85, 89, 91, 87, 93, 84, 97, 95]
}

#DataFrame creation
df=pd.DataFrame(d)

#Creating box plot
plt.boxplot(df.values,labels=df.columns,patch_artist=True)

#grid
plt.grid(True,linestyle='--',alpha=0.7)

#title,ylabel
plt.title("DISTRIBUTION OF SCORES ACROSS SUBJECTS",color="#ff001f")
plt.ylabel("Marks")

#display
plt.show()
```

```python
#5. Histogram of Age Distribution

#Goal: Plot a histogram for ages of people attending an event.

#Extra: Customize bins and add a density curve.

#Skills: Histogram, data distributionÂ understanding.
```

```python
#importing libraries
import matplotlib.pyplot as plt
import numpy as np

#data
ages = [18, 20, 22, 19, 21, 25, 28, 26, 30, 24, 27, 29, 31, 33, 35,
        37, 40, 42, 38, 36, 32, 34, 45, 48, 50, 52, 41, 43, 47, 49,
        22, 24, 26, 28, 29, 31, 33, 35, 38, 39, 44, 46, 51, 55, 60]

#creating histogram
count,bins,_=plt.hist(ages,bins=10,color='#ff99aa',edgecolor='black',density=True)

#Density curve
plt.plot(bins[:-1],count,linewidth=2,color='green')

#ylabel,title
plt.xlabel("Age of Students")
plt.ylabel("Number of Students")
plt.title("AGE DISTRIBUTION OF STUDENTS",color='blue')

#displying
plt.show()
```
