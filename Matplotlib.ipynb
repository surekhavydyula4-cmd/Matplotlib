```python
#line graph
import matplotlib.pyplot as plt
x=[21,34,65]
y=[34,24,15]
plt.plot(x,y)
plt.show()
```

```python
#line graph with color keyword argument
import matplotlib.pyplot as plt
x=[5,7,10]
y=[12,20,34]
plt.plot(x,y,color='g')
plt.show()
```

```python
# line graph with marker keyword argument
import matplotlib.pyplot as plt
x=[12,20,36]
y=[10,22,40]
plt.plot(x,y,marker="*")
plt.show()
```

```python
#line graph with color and marker
import matplotlib.pyplot as plt
x=[12,23,38]
y=[12,53,67]
plt.plot(x,y,color="orange",marker="s")
plt.show()
```

```python
#line graph with linestyle
import matplotlib.pyplot as plt
x=[12,23,38]
y=[12,53,67]
plt.plot(x,y,linestyle="-.")
plt.show()
```

```python
#line graph with color,marker and linestyle
import matplotlib.pyplot as plt
x=[12,23,38]
y=[12,53,67]
plt.plot(x,y,color="#f001ff",marker="^",linestyle="-.")
plt.show()
```

```python
#line graph with color and marker color
import matplotlib.pyplot as plt
x=[12,23,38]
y=[12,53,67]
plt.plot(x,y,color="black",marker="o",linestyle="-.",markerfacecolor="r",markeredgecolor='r')
plt.show()
```

```python
# line graph with title()
import matplotlib.pyplot as plt
x=[5,6,8]
y=[5,10,45]
plt.title("Line Graph")
plt.plot(x,y)
plt.show()
```

```python
#line graph with x and y labels
import matplotlib.pyplot as plt
x=[21,34,25]
y=[12,26,47]
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.plot(x,y)
plt.show()
```

```python
#line graph with x and y labels
import matplotlib.pyplot as plt
x=[21,34,25]
y=[12,26,47]
plt.xlabel("X-axis",color='g')
plt.ylabel("Y-axis",color='g')
plt.title("Line Graph",color='r')
plt.plot(x,y)
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
m=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
s=[120,230,422,87,813,625,541,133,1423,615,452,123]
plt.bar(m,s)
plt.show()
```

```python
#bar graph with title,x and y labels
import matplotlib.pyplot as plt
m=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
s=[120,230,422,87,813,625,541,133,1423,615,452,123]
plt.title("Months Product Sales")
plt.xlabel("Months")
plt.ylabel("Product Sales")
plt.bar(m,s)
plt.show()
```

```python
#bar graph with title,x and y labels ,color
import matplotlib.pyplot as plt
m=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
s=[120,230,422,87,813,625,541,133,1423,615,452,123]
plt.title("Months Product Sales")
plt.xlabel("Months")
plt.ylabel("Product Sales")
plt.bar(m,s,color="yellow",edgecolor="black")
plt.show()
```

```python
# Horizontal bar graph
import matplotlib.pyplot as plt
m=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
s=[120,230,422,87,813,625,541,133,1423,615,452,123]
plt.title("Months Product Sales")
plt.xlabel("Product Sales")
plt.ylabel("Months")
plt.barh(m,s,color="#f000ff",edgecolor="green")
plt.show()
```

```python
#Histogram with title
import matplotlib.pyplot as plt
p=[12,11,13,14,15,16,12,12,13,11,11,15,16,17,14,12]
plt.title("Histogram")
plt.hist(p)
plt.show()
```

```python
#Histogram with color
import matplotlib.pyplot as plt
p=[12,11,13,14,15,16,12,12,13,11,11,15,16,17,14,12]
plt.title("Histogram")
plt.hist(p,color="green",edgecolor="black")
plt.show()
```

```python
#Histogram with color and bins
import matplotlib.pyplot as plt
p=[12,11,13,14,15,16,12,12,13,11,11,15,16,17,14,12]
plt.title("Histogram")
plt.hist(p,bins=15,color="pink",edgecolor="black")
plt.show()
```

```python
#pie chart with labels
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
plt.pie(h,labels=d)
plt.show()
```

```python
#pie chart with color
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c)
plt.show()
```

```python
#pie chart with startangle
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,startangle=270)
plt.show()
```

```python
#pie chart with autopct
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,autopct='%1.1f%%')
plt.show()
```

```python
#pie chart with explode
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,explode=[0.01,0.01,0.01,0.01,0.01],autopct='%1.1f%%')
plt.show()
```

```python
#pie chart with shadow
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,shadow=True,explode=[0.01,0.01,0.01,0.01,0.01],autopct='%1.1f%%')
plt.show()
```

```python
#pie chart with axis
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,shadow=True,explode=[0.01,0.01,0.01,0.01,0.01],autopct='%1.1f%%')
plt.axis('Equal')
plt.show()
```

```python
#pie chart with legend()
import matplotlib.pyplot as plt
d=["Sleep","Study","Entertainment","Travel","Others"]
h=[8,4,4,2,6]
c=['g','purple','r','Yellow','b']
plt.pie(h,labels=d,colors=c,shadow=True,explode=[0.05,0.05,0.05,0.05,0.05],autopct='%1.1f%%')
plt.axis('Equal')
plt.legend()
plt.show()
```
