```python
#scatter plot
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
r=[10,20,30,40,50]
plt.xlabel("Area")
plt.ylabel("Rice-Bags")
plt.title("Rice-Bags in Every Area")
plt.scatter(a,r)
plt.show()
```

```python
#heatmap
import matplotlib.pyplot as plt
import pandas as pd
d={
    "Apples":[10,20,33,45],
    "Mango":[12,23,14,26],
    "Grapes":[23,24,34,25],
    "Guavas":[21,1,14,15]
    }
i=["Basket1","Basket2","Basket3","Basket4"]
df=pd.DataFrame(d,index=i)
print(df)
plt.imshow(df)
plt.colorbar()
plt.show()
```

```python
#heatmap with x and y ticks
import matplotlib.pyplot as plt
import pandas as pd
d={
    "Apples":[10,20,33,45],
    "Mango":[12,23,14,26],
    "Grapes":[23,24,34,25],
    "Guavas":[21,1,14,15]
    }
i=["Basket1","Basket2","Basket3","Basket4"]
df=pd.DataFrame(d,index=i)
print(df)
plt.imshow(df)
plt.colorbar()
plt.xticks(range(len(df)),df.columns)
plt.yticks(range(len(df)),df.index)
plt.show()
```

```python
#heatmap with colormap
import matplotlib.pyplot as plt
import pandas as pd
d={
    "Apples":[10,20,33,45],
    "Mango":[12,23,14,26],
    "Grapes":[23,24,34,25],
    "Guavas":[21,1,14,15]
    }
i=["Basket1","Basket2","Basket3","Basket4"]
df=pd.DataFrame(d,index=i)
print(df)
plt.imshow(df,cmap = "YlGnBu")
plt.colorbar()
plt.xticks(range(len(df)),df.columns)
plt.yticks(range(len(df)),df.index)
plt.show()
```
